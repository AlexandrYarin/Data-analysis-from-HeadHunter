3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(string):
    if string is np.nan or string == 'Не указано':
        return np.nan
    year_words=['год', 'года', 'лет']
    month_words=['месяц', 'месяца', 'месяцев']
    string_splitted = string.split(' ')[:7]
    years = 0
    months = 0
    for index, item in enumerate (string_splitted):
        if item in year_words:
            years = int(string_splitted[index-1])
        if item in month_words:
            months = int(string_splitted[index-1])
    return int(years*12 + months)
        
hh_data['Опыт работы(месяц)'] = hh_data['Опыт работы'].apply(get_experience)
hh_data.drop('Опыт работы', axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city(string):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                      'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    capitals = ['Москва', 'Санкт-Петербург']
    x = string.split()[0]
    if x in capitals:
        return x
    elif x in million_cities:
        return 'город-миллионик'
    else:
        return 'другие'

def get_moving(string):
    x = string.split(',')
    if len(x) != 4:
        goal = x[1].strip()
        return False if goal.lower().startswith('не') else True
    else:
        goal = x[2].strip()
        return False if goal.lower().startswith('не') else True

def get_business_trip(string):
    if 'командировка' in string:
        if ('не готов к командировкам' in string) or ('не готова к командировкам' in string):
            return False
        else: 
            return True
    else:
        return False

hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_city)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(get_moving)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(get_business_trip)
```

```python
exchange = pd.read_csv('C:/Users/marcu/OneDrive/Рабочий стол/YA/Lessons/project_1/data/ExchangeRates.csv')
exchange.drop(['per', 'time', 'vol'], axis=1, inplace=True)
```

```python
currency_dict = {
'руб': 'RUB',
'грн':'UAH',
'USD':'USD',	
'EUR':'EUR',	
'белруб':'BYN',
'KGS':'KGS',
'сум':'UZS',	
'AZN':'AZN',	
'KZT':'KZT'
}

exchange['date'] = pd.to_datetime(exchange['date'], dayfirst=True).dt.date
hh_data['date'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date
hh_data['currency'] = hh_data['ЗП'].apply(lambda x: currency_dict[x.split()[1].replace('.', '')])
hh_data['value'] = hh_data['ЗП'].apply(lambda x: int(x.split()[0]))
```

```python
hh_data = hh_data.merge(exchange, how='left', on=['currency', 'date'])
[hh_data[col].fillna(1, inplace=True) for col in ['close', 'proportion']]
hh_data['ЗП (руб)'] = (hh_data['value'] * hh_data['close']) / hh_data['proportion']
hh_data.drop(['ЗП', 'value'],axis=1, inplace=True)
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
log_age = np.log(hh_data['Возраст'])
mu, sigma = log_age.mean(), log_age.std()
fig = plt.figure(figsize=(12, 4))

hist = sns.histplot(log_age, bins=35)
hist.axvline(mu + 3 * sigma, color='k', lw=2, ls='--')
hist.axvline(mu - 3 * sigma, color='k', lw=2, ls='--')
hist.axvline(mu, lw=2, color='k')
hist.set_title('Распределение возраста');
```

```python
print(log_age.skew())

def z_mode(data, features, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[features])
    else:
        x= data[features]
    mu, sigma = x.mean(), x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    
    return outliers, cleaned

outliers, _ = z_mode(hh_data, 'Возраст', right=4, log_scale=True)

print(list(outliers["Возраст"]))
print(outliers.shape[0])
```
